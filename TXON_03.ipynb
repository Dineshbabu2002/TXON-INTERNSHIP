```python
import scipy.stats as stats #probability distributions functions
```

```python
for i,j in list(zip(df.drop('target',axis=1).columns,df.drop('target',axis=1).skew())):
    if df[i].skew()<-3 or df[i].skew()>3:
        df[i] = pd.Series(stats.yeojohnson(df[i])[0])
    else:
        df[i] = df[i]*1
```

```python
from sklearn.feature_selection import SequentialFeatureSelector as SFS,RFE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr=LogisticRegression()
rfe = RFE(estimator=lr,n_features_to_select=35)
rfe_model = rfe.fit(xtrain,ytrain)
rfe_model.ranking_
```

```python
rank_df=pd.DataFrame({'features':xtrain.columns,'ranks':rfe_model.ranking_})
rank_df.head()
```

```python
feat=list(rank_df[rank_df['ranks']==1]['features'].values)
feat
```

```python
x = x[feat]
x.head()
```

```python
df['x38'].unique()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
lr_model = lr.fit(xtrain,ytrain)
lr_model
```

```python
y_pred_train = lr_model.predict(xtrain)
y_pred_test = lr_model.predict(xtest)
```

```python
accuracy_score(ytrain,y_pred_train),accuracy_score(ytest,y_pred_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_model = dt.fit(xtrain,ytrain)
dt_model
```

```python
y_pred_train = dt_model.predict(xtrain)
y_pred_test = dt_model.predict(xtest)
```

```python
accuracy_score(ytrain,y_pred_train),accuracy_score(ytest,y_pred_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf_model = rf.fit(xtrain,ytrain)
rf_model
```

```python
y_pred_train = rf_model.predict(xtrain)
y_pred_test = rf_model.predict(xtest)
```

```python
accuracy_score(ytrain,y_pred_train),accuracy_score(ytest,y_pred_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
y_pred_train = mlp_model.predict(xtrain)
y_pred_test = mlp_model.predict(xtest)
```
