```python
# Group the data by the 'Species' column and count occurrences
species_count = (
    df.groupby('Species')  # Group by 'Species' column
      .size()              # Get the size of each group
      .to_frame('count')   # Convert to DataFrame and name the count column
      .reset_index()       # Reset the index to make 'Species' a column
)

# Display the resulting DataFrame
print(species_count)
```

```python
#bar plot
sns.countplot(x='Species', data=df)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data = df, kind='reg')
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data = df, kind='hex')
plt.show()
```

```python
fg = sns.FacetGrid(df, hue='Species')
fg.map(plt.scatter, 'SepalLengthCm','SepalWidthCm')
fg.add_legend()
plt.show()
```

```python
df.head()
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig = sns.boxenplot(x='Species', y='PetalLengthCm', data=df, order=['Iris-virginica','Iris-versicolor','Iris-setosa'], linewidth=2.5,orient='v',dodge=False)
plt.show()
```

```python
df.boxplot(by='Species', figsize=(14, 7))
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(9,8)
fig = sns.stripplot(x='Species', y='SepalLengthCm', data=df, jitter=True, edgecolor='gray', size=10, palette='winter', orient='v')
plt.show()
```

```python
# Combining Box and Strip Plots
fig = plt.gcf()
fig.set_size_inches(10, 7)
fig = sns.boxplot(x='Species', y='SepalLengthCm', data=df)
fig=sns.stripplot(x='Species', y='SepalLengthCm', data=df, jitter=True, edgecolor='red')
plt.show()
```

```python
ax = sns.boxplot(x='Species', y='SepalLengthCm', data=df)
ax = sns.stripplot(x='Species', y='SepalLengthCm', data=df, jitter=True, edgecolor="gray")

if len(ax.artists) > 2:
    boxtwo = ax.artists[2]
    boxtwo.set_facecolor('yellow')
    boxtwo.set_edgecolor('black')
    
    boxthree = ax.artists[1]
    boxthree.set_facecolor('red')
    boxthree.set_edgecolor('black')
    
    boxone = ax.artists[0]
    boxone.set_facecolor('green')
    boxone.set_edgecolor('black')
plt.show()
```

```python
# Violin plot
fig = plt.gcf()
fig.set_size_inches(12,8)
fig = sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2,2,1)
sns.violinplot(x='Species', y='PetalLengthCm', data=df)
plt.title("petal Leangth")
plt.subplot(2,2,1)
sns.violinplot(x='Species', y='PetalWidthCm', data=df)
plt.title("PetalWidthCm")
plt.subplot(2,2,1)
sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.title("SepalLengthCm")
plt.subplot(2,2,1)
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title("SepalWidthCm")
plt.tight_layout()
plt.show()
```

```python
# pair plot
sns.pairplot(data=df, kind='scatter')
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.show()
```

```python
# Heat Map
plt.figure(figsize=(10, 7))  # Set the figure size
sns.heatmap(
    df.corr(numeric_only=True), 
    annot=True, 
    cmap='cubehelix', 
    linewidths=1, 
    linecolor='k', 
    square=True, 
    vmin=-1, 
    vmax=1, 
    cbar_kws={"orientation": "vertical"}
)
plt.show()
```

```python
# Distribution plot
df.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```
